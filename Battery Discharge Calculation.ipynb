```python
from IPython.display import clear_output

a1 = float(input("Actual Discharge Duration(Hours): "))
a2 = float(input("Rated Battery Capacity(Ah): "))
a3 = float(input("Load Current per Charger(A): "))
clear_output()
d1 = float(a1/5)

if a1<=5:
    print("Discharge Test Ratio:",d1)
else:
    print("Wrong Input")
    
s1 = (a2/a3)*d1
print("Standby Time",s1)

if d1<0.80 and s1 >=8:
    print("Battery bank is unhealthy but standby time is acceptable.\n")
    print("Action:")
    print("1.Replace bank battery once average conductance is equal or below 60%.")
    print("2.Replaced battery bank to be scrapped.")
if d1<0.8 and s1<8:
    print("Battery bank is unhealthy and standby time is unacceptable.\n")
    print("Action:")
    print("1.Replace battery bank immediately.")
    print("2.Replaced battery bank to be scrapped.")
if d1 >=0.8 and s1>=8:
    print("Battery bank is healthy and standby time is acceptable.\n")
    print("Action:")
    print("1.Continue battery bank maintenance as per the maintenance schedule.")
if d1 >=0.8 and s1<8:
    print("Battery bank is healthy but standby time unacceptable.\n")
    print("Action:")
    print("1.Replace battery bank with higher capacity battery.")
    print("2.Replaced battery bank to be relocated.")
    
```
